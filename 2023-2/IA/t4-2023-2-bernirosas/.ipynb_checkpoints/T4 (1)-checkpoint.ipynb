## 1.3 Vectorizaci√≥n (0.9 ptos)

En esta secci ÃÅon, es fundamental que transformes la informaci√≥n de las rese√±as escritas en vectores num√©ricos que seran utilizados para entrenar los modelos. Para lograr esto, deber√°s hacer uso de Bag of Words, Tf-Idf y SBERT.

### 1.3.1 Bag of Words

Investiga la vectorizaci√≥n Bag of Words, explica su funcionamiento y apl√≠cala sobre los datos procesados, luego almacena el resultado en un DataFrame llamado X_bow.

```python
data_a_usar.reset_index(drop=True, inplace=True)
cuerpo = data_a_usar['review']

vectorizer = CountVectorizer()

X_bow_vector = vectorizer.fit_transform(cuerpo)

X_bow = pd.DataFrame(X_bow_vector.toarray(), columns=vectorizer.get_feature_names_out())

```

```python
print(vectorizer.vocabulary_)
```

```python
X_bow
```

### 1.3.2 TF-IDF

Investiga la vectorizaci√≥n TF-IDF, explica su funcionamiento y apl√≠cala sobre los datos procesados, luego almacena el resultado en un DataFrame llamado X_tfidf.

```python
data_a_usar.reset_index(drop=True, inplace=True)
cuerpo = data_a_usar['review']

TF_vectorizer = TfidfVectorizer()

TF_vector = TF_vectorizer.fit_transform(cuerpo)

X_tfidf = pd.DataFrame(TF_vector.toarray(), columns=TF_vectorizer.get_feature_names_out())
```

```python
print(sorted(TF_vectorizer.vocabulary_))
```

### 1.3.3 SBERT

Investiga la vectorizaci√≥n SBERT, explica su funcionamiento y apl√≠cala a los datos procesados. Luego, almacena el resultado en un DataFrame llamado X_sbert.

**[IMPORTANTE:]**
Para transformar texto a vectores utilizando esta estrategia, el tiempo que lleva procesar una cantidad grande de texto puede ser mucha. Si se est√° tardando mucho, puedes disminuir la cantidad de datos.

```python
filas_positivas = data_a_usar[data_a_usar['sentiment'] == 'positive']
muestra_positiva = filas_positivas.sample(n=1000, random_state=20)
filas_negativas = data_a_usar[data_a_usar['sentiment'] == 'negative']
muestra_negativa = filas_negativas.sample(n=1000, random_state=20)

data_a_usar_SBERT = pd.concat([muestra_negativa, muestra_positiva])
data_a_usar_SBERT['sentiment'].value_counts()
data_a_usar_SBERT = data_a_usar_SBERT.sample(n=2000, random_state=20) # para que no esten ordenados

data_a_usar_SBERT.reset_index(drop=True, inplace=True)
cuerpo = data_a_usar_SBERT['review']
```

```python
data_a_usar_SBERT
```

```python
# c√≥digo en base a https://anirbansen2709.medium.com/sbert-how-to-use-sentence-embeddings-to-solve-real-world-problems-f950aa300c72
model = SentenceTransformer('all-MiniLM-L6-v2')

X_sbert_vector = model.encode(cuerpo, convert_to_tensor=True)

X_sbert = pd.DataFrame(X_sbert_vector.numpy())
```

```python
X_sbert
```

## 1.4 Utilizaci√≥n de Clasificadores (0.9 ptos)

### 1.4.1 √Årbol de Decisi√≥n

- √Årbol de Decisi√≥n con Bag of Words

```python
from sklearn.metrics import accuracy_score
# codigo desde ayudant√≠a
y = data_a_usar['sentiment']

X_train_tree_bow, X_test, y_train_tree_bow, y_test = train_test_split(X_bow, y, test_size=0.2, random_state=40)

clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train_tree_bow, y_train_tree_bow)

y_pred = clf.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy del modelo:", accuracy)

conf_matrix = confusion_matrix(y_test, y_pred)
print("Matriz de Confusi√≥n:")
print(conf_matrix)

class_report = classification_report(y_test, y_pred)
print("Informe de Clasificaci√≥n:")
print(class_report)
```

- √Årbol de Decisi√≥n con SBERT


```python
from sklearn.metrics import accuracy_score
y = data_a_usar_SBERT['sentiment']

X_train, X_test, y_train, y_test = train_test_split(X_sbert, y, test_size=0.2, random_state=40)

clf_tree_sbert = DecisionTreeClassifier(random_state=42)
clf_tree_sbert.fit(X_train, y_train)

y_pred = clf_tree_sbert.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Precisi√≥n del modelo:", accuracy)

conf_matrix = confusion_matrix(y_test, y_pred)
print("Matriz de Confusi√≥n:")
print(conf_matrix)

class_report = classification_report(y_test, y_pred)
print("Informe de Clasificaci√≥n:")
print(class_report)
```

- √Årbol de Decisi√≥n con TF-IDF

```python
from sklearn.metrics import accuracy_score
y = data_a_usar['sentiment']

X_train, X_test, y_train, y_test = train_test_split(X_tfidf, y, test_size=0.2, random_state=40)

clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Precisi√≥n del modelo:", accuracy)

conf_matrix = confusion_matrix(y_test, y_pred)
print("Matriz de Confusi√≥n:")
print(conf_matrix)

class_report = classification_report(y_test, y_pred)
print("Informe de Clasificaci√≥n:")
print(class_report)
```

### 1.4.3 Random Forest
